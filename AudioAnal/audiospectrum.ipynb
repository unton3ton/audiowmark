```python
# https://www.pythonpool.com/spectrogram-python/
import os

import matplotlib.pyplot as plt
%matplotlib inline

import librosa
import librosa.display
 
import IPython.display as ipd
```

```python
audio = '611.wav'
x, sr = librosa.load(audio, sr=44100)
```

```python
# https://stackoverflow.com/questions/75799092/librosa-error-attributeerror-module-librosa-display-has-no-attribute-wave
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr) 
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
audio1 = '611wm.wav'
x1, sr1 = librosa.load(audio1, sr=44100)
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x1, sr=sr1)
```

```python
X1 = librosa.stft(x1)
Xdb1 = librosa.amplitude_to_db(abs(X1))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb1, sr=sr1, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x1-x, sr=sr1)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb1-Xdb, sr=sr1, x_axis='time', y_axis='hz')
plt.colorbar()
```
